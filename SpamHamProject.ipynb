```python
import numpy as np 
import pandas as pd 
```

```python
messages=pd.read_csv('SMSSpamCollection.csv',sep='\t',names=['label','message'])
```

```python
messages
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus =[]
for i in range(len(messages)):
    review=re.sub('[^a-zA-Z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if word  not in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
import numpy as np
np.set_printoptions(edgeitems=30,linewidth=100000,formatter=dict(float=lambda x:"%.3g" %  x))
```

```python
## before doing the Bag Of Words here we are doing TRain test split so that data leakage should not occur.
## output feature
y=pd.get_dummies(messages['label'])
y=y.iloc[:,0].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(corpus,y,test_size=0.30,random_state=42)
```

## Create Bag Of Words

```python
## create the bag of words model.
from sklearn.feature_extraction.text import CountVectorizer
## for binary bow enable binary = true.
cv=CountVectorizer()
```

```python
X_train ## my train data is still a text data
```

```python
X_train=cv.fit_transform(X_train).toarray() ## converting the training data into vector
```

```python
X_test
```

```python
X_test=cv.transform(X_test).toarray()
```

```python
import numpy as np
np.set_printoptions(edgeitems=30,linewidth=100000,formatter=dict(float=lambda x:"%.3g" %  x))
```

```python
cv.vocabulary_
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_detect_model=MultinomialNB().fit(X_train,y_train)
```

```python
spam_detect_model
```

```python
y_pred=spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
acc_score=accuracy_score(y_test,y_pred)
cr=classification_report(y_test,y_pred)
```

```python
print(acc_score)
print(cr)
```
